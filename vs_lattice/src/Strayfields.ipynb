```julia
using Plots
using LinearAlgebra
include("Unitcell.jl")
include("DW_Lattice.jl")
```

```julia
nm = 1E-9;
taylorHz(y,x) = clamp(-∂Hz_∂x(AFM2R_lattice,y*nm)*x*nm,-1000,1000)
#taylorHz(y,x) = y^2*x/nm^3
yvals = collect(LinRange(-100,100,300)); xvals = collect(LinRange(-30,30,100))
taylorHz(10*nm,30*nm)
```

```julia
contourf(yvals,xvals,real∘taylorHz,clabels=true,thickness_scaling=1.5,
        cmap=:turbo,levels=collect([sign(i)*i^2 for i = -60:5:60]),dpi=600,lw=2,size=(700,500),xlabel="Y position (nm)", ylabel="X position along RT (nm)")
```
